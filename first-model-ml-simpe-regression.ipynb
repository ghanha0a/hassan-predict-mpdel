```python
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
x= data ['YearsExperience'].values
y= data['Salary'].values
```

```python
plt.scatter(x,y)
plt.show()
```

**#start train my ML model**# 

```python
from sklearn.linear_model import LinearRegression
```

trainig

```python
x= x.reshape(-1, 1)
y= y.reshape(-1, 1)
```

```python
model= LinearRegression()
model.fit(x,y)
```

acurrcy 

```python
model.score(x,y)
```

predict

```python
model.predict([[2.1]])
```
